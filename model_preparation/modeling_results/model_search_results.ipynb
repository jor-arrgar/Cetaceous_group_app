Due to a writting error in the saving .csv code in the model params search notebook, the index needs to be extract manually.

```python
index_100 = []

pool_sizes_1 = [(3,3) , (5,5)]
pool_sizes_2 = [(3,3) , (5,5)]
pool_sizes_3 = [(3,3) , (5,5)]

n_neurons = [400, 500, 600]

epochs = [20 , 30 , 40 , 50]
batchs = [32 , 64 , 128]



for pool_1 in pool_sizes_1:
    for pool_2 in pool_sizes_2:
        for pool_3 in pool_sizes_3:
            for n in n_neurons:
                for epoch in epochs:
                    for batch in batchs:
                        index_100.append((100 , pool_1 , pool_2 , pool_3 , n , epoch , batch))

index_100 = index_100[:42]

model_index = pd.MultiIndex.from_tuples(index_100)
df_100_pixels =  df_100_pixels.set_index(model_index)
df_100_pixels.head()
```

```python
df_100_pixels.tail()
```

```python
index_32 = []

pool_sizes_1 = [(2,2) , (3,3) , (4,4) , (5,5)]
pool_sizes_2 = [(2,2) , (3,3) , (4,4) , (5,5)]
pool_sizes_3 = [(2,2) , (3,3) , (4,4) , (5,5)]

n_neurons = [400, 500, 600]

epochs = [20 , 30 , 40 , 50]
batchs = [32 , 64 , 128]

for pool_1 in pool_sizes_1:
    for pool_2 in pool_sizes_2:
        for pool_3 in pool_sizes_3:
            for n in n_neurons:
                for epoch in epochs:
                    for batch in batchs:
                        index_32.append((32 , pool_1 , pool_2 , pool_3 , n , epoch , batch))

index_32 = index_32[:108]

model_index = pd.MultiIndex.from_tuples(index_32)
df_32_pixels =  df_32_pixels.set_index(model_index)
df_32_pixels.head()
```

The parameters selected for the prediction model are the following:

- Image shape (pixels) ==> 100 x 100
- Image format (color) ==> Full color (3 layers RGB)

- 1st convolutional layer (filters, strides) == (32 , (3,3)) ----- Constant during experiment
- 1st pooling layer (size) ==> (3,3)
- 2nd convolutional layer (filters, strides) == (64 , (3,3)) ----- Constant during experiment
- 2nd pooling layer (size) ==> (3,3)
- 3rd convolutional layer (filters, strides) == (128 , (3,3)) ----- Constant during experiment
- 3rd pooling layer (size) ==> (3,3)
- Activation functions (for all layers) ==> 'relu' ----- Constant during experiment
- Dropout (for all convolutional layers) ==> 0.25 ----- Constant during experiment

- Flatten layer
- 1st dense layer (n_neurons , activation , dropout) ==> 400 , 'relu' , 0.5
- Exit dense layer (n_neurons , activarion) ==> 3 , 'softmax'

- Compile parameters (optimizer , loss , metrics) ==> 'adam' , 'categorical_crossentropy' , 'accuracy' ----- Constant during experiment

- Training parameters (epochs, batch size) ==> 50 , 128




```python
model_1 = [('32x32'),('2 x 2'),('2 x 2'),('4 x 4'),'400','30','128','0.048675','0.983000','0.871667','0.539551','0.885556']
model_2 = [('32x32'),('2 x 2'),('2 x 2'),('4 x 4'),'500','30','128','0.058405','0.978333','0.865000','0.481576','0.885556']
model_3 = [('100 x 100'),('3 x 3'),('3 x 3'),('3 x 3'),'400','50','128','0.009912','0.996667','0.885','0.583434','0.912222']
```

```python
params_values = np.array([model_1 , model_2 , model_3])
params_names = ['Image size (pixels)' , '1st pooling layer (size)' , '2nd pooling layer (size)' , '3rd pooling layer(size)' , '1st dense layer neurons' , 'Epochs' , 'Batch size' , 'Train loss' , 'Train accuracy' ,	'Validation accuracy' , 'Test loss' , 'Test accuracy']
params_df = pd.DataFrame(params_values , columns= params_names)
params_df
```

```python
params_df.to_csv('../../cetaceous_group_predictor/best_models.csv' , sep= ';' , index=False)
```
